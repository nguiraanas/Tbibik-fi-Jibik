```python
"""
================================================================================
FUSION MULTIMODALE - PARTIE 1/2 - PHASE 1
================================================================================
Cette partie contient:
- Architectures compl√®tes
- Phase 1 d'entra√Ænement (10 epochs)
- Sauvegarde du meilleur mod√®le

√Ä T√âL√âCHARGER APR√àS EX√âCUTION:
- best_fusion_phase1_final.pth (meilleur mod√®le Phase 1)
- fusion_phase1_history.json (historique d'entra√Ænement)
- fusion_phase1_config.json (configuration pour Phase 2)
================================================================================
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts
import torchvision.models.video as video_models
from sklearn.metrics import (accuracy_score, f1_score, confusion_matrix,
                             roc_auc_score)
import time
import json
import os
import zipfile
from tqdm.auto import tqdm
import warnings
warnings.filterwarnings('ignore')

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("="*80)
print("üîÄ FUSION MULTIMODALE - PARTIE 1/2 - PHASE 1")
print("="*80)
print(f"üñ•Ô∏è  Device: {device}\n")

# ============================================================================
# 1. ARCHITECTURES ORIGINALES
# ============================================================================

class SpatialAttention(nn.Module):
    def __init__(self, in_channels):
        super().__init__()
        self.conv = nn.Conv3d(in_channels + 1, 1, kernel_size=1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, features, masks):
        if masks.shape[-2:] != features.shape[-2:]:
            masks = F.interpolate(
                masks.view(masks.size(0), 1, -1, masks.size(-2), masks.size(-1)),
                size=features.shape[-3:], mode='trilinear', align_corners=False
            )
        combined = torch.cat([features, masks], dim=1)
        attention = self.sigmoid(self.conv(combined))
        return features * attention


class VideoModelWithSegmentation(nn.Module):
    def __init__(self, num_classes=2, pretrained=True, dropout=0.5):
        super().__init__()
        self.backbone = video_models.r2plus1d_18(pretrained=pretrained)
        self.backbone.stem[0] = nn.Conv3d(1, 45, kernel_size=(1,7,7),
                                          stride=(1,2,2), padding=(0,3,3), bias=False)
        self.attention = SpatialAttention(in_channels=64)
        in_features = self.backbone.fc.in_features
        self.backbone.fc = nn.Identity()
        
        self.classifier = nn.Sequential(
            nn.Linear(in_features, 256), nn.ReLU(), nn.BatchNorm1d(256), nn.Dropout(dropout),
            nn.Linear(256, 128), nn.ReLU(), nn.BatchNorm1d(128), nn.Dropout(dropout*0.5),
            nn.Linear(128, num_classes)
        )
        
        self.intermediate_features = None
        def hook_fn(module, input, output):
            self.intermediate_features = output
        self.backbone.layer1.register_forward_hook(hook_fn)
    
    def forward(self, video, masks, return_features=False):
        x = video.permute(0, 2, 1, 3, 4)
        masks = masks.permute(0, 2, 1, 3, 4)
        x = self.backbone.stem(x)
        x = self.backbone.layer1(x)
        x = self.attention(x, masks)
        x = self.backbone.layer2(x)
        x = self.backbone.layer3(x)
        x = self.backbone.layer4(x)
        x = self.backbone.avgpool(x)
        features = x.flatten(1)
        logits = self.classifier(features)
        if return_features:
            return logits, features
        return logits


class EFBranch(nn.Module):
    def __init__(self, num_classes=2, dropout=0.5):
        super().__init__()
        self.ef_branch = nn.Sequential(
            nn.Linear(1, 64), nn.ReLU(), nn.BatchNorm1d(64), nn.Dropout(dropout),
            nn.Linear(64, 128), nn.ReLU(), nn.BatchNorm1d(128), nn.Dropout(dropout),
            nn.Linear(128, 64), nn.ReLU(), nn.BatchNorm1d(64), nn.Dropout(dropout*0.5),
            nn.Linear(64, num_classes)
        )
    
    def forward(self, ef, return_features=False):
        ef = ef.unsqueeze(1)
        if return_features:
            x = ef
            for i, layer in enumerate(self.ef_branch):
                x = layer(x)
                if i == len(self.ef_branch) - 2:
                    features = x
            logits = self.ef_branch[-1](features)
            return logits, features
        return self.ef_branch(ef)


print("‚úÖ Architectures originales d√©finies\n")

# ============================================================================
# 2. CROSS-MODAL ATTENTION AM√âLIOR√âE (SOFT REGULARIZATION)
# ============================================================================

class SoftBalancedCrossModalAttention(nn.Module):
    """Attention SANS clipping - apprentissage libre avec soft regularization"""
    def __init__(self, video_dim=512, ef_dim=64, hidden_dim=128):
        super().__init__()
        self.video_proj = nn.Linear(video_dim, hidden_dim)
        self.ef_proj = nn.Linear(ef_dim, hidden_dim)
        self.attention_score = nn.Linear(hidden_dim, 1)
    
    def forward(self, video_features, ef_features):
        video_proj = self.video_proj(video_features)
        ef_proj = self.ef_proj(ef_features)
        
        video_score = self.attention_score(torch.tanh(video_proj))
        ef_score = self.attention_score(torch.tanh(ef_proj))
        
        scores = torch.cat([video_score, ef_score], dim=1)
        attention_weights = F.softmax(scores, dim=1)
        
        attended_video = video_features * attention_weights[:, 0:1]
        attended_ef = ef_features * attention_weights[:, 1:2]
        
        return attended_video, attended_ef, attention_weights


class SoftAttentionRegularizationLoss(nn.Module):
    """R√©gularisation L2 douce"""
    def __init__(self, target_balance=0.5, lambda_reg=0.3):
        super().__init__()
        self.target_balance = target_balance
        self.lambda_reg = lambda_reg
    
    def forward(self, attention_weights):
        video_weight = attention_weights[:, 0].mean()
        ef_weight = attention_weights[:, 1].mean()
        
        video_penalty = (video_weight - self.target_balance) ** 2
        ef_penalty = (ef_weight - self.target_balance) ** 2
        
        reg_loss = self.lambda_reg * (video_penalty + ef_penalty)
        return reg_loss


print("‚úÖ Cross-Modal Attention (soft) cr√©√©e\n")

# ============================================================================
# 3. MOD√àLE DE FUSION
# ============================================================================

class OptimizedMultimodalFusion(nn.Module):
    def __init__(self, video_model, ef_model, num_classes=2, dropout=0.5):
        super().__init__()
        self.video_branch = video_model
        self.ef_branch = ef_model
        self.cross_attention = SoftBalancedCrossModalAttention(512, 64, 128)
        
        self.fusion_classifier = nn.Sequential(
            nn.Linear(576, 256), nn.ReLU(), nn.BatchNorm1d(256), nn.Dropout(dropout),
            nn.Linear(256, 128), nn.ReLU(), nn.BatchNorm1d(128), nn.Dropout(dropout*0.5),
            nn.Linear(128, 64), nn.ReLU(), nn.BatchNorm1d(64), nn.Dropout(dropout*0.25),
            nn.Linear(64, num_classes)
        )
    
    def forward(self, video, masks, ef):
        video_logits, video_features = self.video_branch(video, masks, return_features=True)
        ef_logits, ef_features = self.ef_branch(ef, return_features=True)
        
        attended_video, attended_ef, attention_weights = self.cross_attention(
            video_features, ef_features
        )
        
        fused_features = torch.cat([attended_video, attended_ef], dim=1)
        fusion_logits = self.fusion_classifier(fused_features)
        
        return fusion_logits, video_logits, ef_logits, attention_weights
    
    def freeze_branches(self):
        for param in self.video_branch.parameters():
            param.requires_grad = False
        for param in self.ef_branch.parameters():
            param.requires_grad = False


print("‚úÖ Mod√®le de fusion cr√©√©\n")

# ============================================================================
# 4. LOSS AVEC R√âGULARISATION PROGRESSIVE
# ============================================================================

class ProgressiveMultimodalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0, fusion_weight=0.7, 
                 branch_weight=0.15, lambda_attn_reg=0.5):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.fusion_weight = fusion_weight
        self.branch_weight = branch_weight
        self.attn_reg = SoftAttentionRegularizationLoss(
            target_balance=0.5, lambda_reg=lambda_attn_reg
        )
    
    def focal_loss(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, weight=self.alpha, reduction='none')
        pt = torch.exp(-ce_loss)
        return ((1 - pt) ** self.gamma * ce_loss).mean()
    
    def forward(self, fusion_logits, video_logits, ef_logits, targets, attention_weights):
        fusion_loss = self.focal_loss(fusion_logits, targets)
        video_loss = self.focal_loss(video_logits, targets)
        ef_loss = self.focal_loss(ef_logits, targets)
        attn_reg_loss = self.attn_reg(attention_weights)
        
        total = (self.fusion_weight * fusion_loss +
                 self.branch_weight * video_loss +
                 self.branch_weight * ef_loss +
                 attn_reg_loss)
        
        return total, {
            'total': total.item(),
            'fusion': fusion_loss.item(),
            'video': video_loss.item(),
            'ef': ef_loss.item(),
            'attn_reg': attn_reg_loss.item()
        }


print("‚úÖ Loss progressive cr√©√©e\n")

# ============================================================================
# 5. FONCTIONS D'ENTRA√éNEMENT
# ============================================================================

def train_epoch(model, loader, criterion, optimizer, device, threshold=0.25):
    model.train()
    losses = {'total': 0.0, 'fusion': 0.0, 'video': 0.0, 'ef': 0.0, 'attn_reg': 0.0}
    all_preds, all_labels = [], []
    all_attention = []
    
    for batch in tqdm(loader, desc='Training', leave=False):
        videos = batch['video'].to(device)
        masks = batch['masks'].to(device)
        ef = batch['ef'].to(device)
        labels = batch['label'].to(device)
        
        optimizer.zero_grad()
        fusion_logits, video_logits, ef_logits, attn = model(videos, masks, ef)
        loss, loss_dict = criterion(fusion_logits, video_logits, ef_logits, labels, attn)
        
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        
        for k in losses:
            losses[k] += loss_dict[k]
        
        probs = torch.softmax(fusion_logits, dim=1)
        preds = (probs[:, 1] > threshold).long()
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_attention.append(attn.detach().cpu())
        
        del videos, masks, ef, labels, fusion_logits, video_logits, ef_logits, loss
        torch.cuda.empty_cache()
    
    epoch_losses = {k: v/len(loader) for k, v in losses.items()}
    acc = accuracy_score(all_labels, all_preds)
    avg_attn = torch.cat(all_attention, dim=0).mean(dim=0)
    return epoch_losses, acc, avg_attn


def validate_epoch(model, loader, criterion, device, threshold=0.15):
    model.eval()
    losses = {'total': 0.0, 'fusion': 0.0, 'video': 0.0, 'ef': 0.0, 'attn_reg': 0.0}
    all_preds, all_labels, all_probs = [], [], []
    all_attention = []
    
    with torch.no_grad():
        for batch in tqdm(loader, desc='Validation', leave=False):
            videos = batch['video'].to(device)
            masks = batch['masks'].to(device)
            ef = batch['ef'].to(device)
            labels = batch['label'].to(device)
            
            fusion_logits, video_logits, ef_logits, attn = model(videos, masks, ef)
            loss, loss_dict = criterion(fusion_logits, video_logits, ef_logits, labels, attn)
            
            for k in losses:
                losses[k] += loss_dict[k]
            
            probs = torch.softmax(fusion_logits, dim=1)
            preds = (probs[:, 1] > threshold).long()
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(probs[:, 1].cpu().numpy())
            all_attention.append(attn.cpu())
            
            del videos, masks, ef, labels
            torch.cuda.empty_cache()
    
    epoch_losses = {k: v/len(loader) for k, v in losses.items()}
    acc = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='weighted')
    auc = roc_auc_score(all_labels, all_probs)
    cm = confusion_matrix(all_labels, all_preds)
    recall_unhealthy = cm[0,0] / (cm[0,0] + cm[0,1]) if (cm[0,0] + cm[0,1]) > 0 else 0
    recall_healthy = cm[1,1] / (cm[1,0] + cm[1,1]) if (cm[1,0] + cm[1,1]) > 0 else 0
    avg_attn = torch.cat(all_attention, dim=0).mean(dim=0)
    
    return epoch_losses, acc, f1, auc, recall_unhealthy, recall_healthy, avg_attn


# ============================================================================
# 6. PHASE 1 D'ENTRA√éNEMENT
# ============================================================================

def train_phase1(train_loader, val_loader, device, config):
    print("\n" + "="*80)
    print("üî• PHASE 1: FINE-TUNING FUSION AVEC R√âGULARISATION PROGRESSIVE")
    print("="*80 + "\n")
    
    # Chargement des mod√®les
    print("üì• Chargement des mod√®les pr√©-entra√Æn√©s...")
    video_model = VideoModelWithSegmentation(num_classes=2, pretrained=False, dropout=0.5)
    ef_model = EFBranch(num_classes=2, dropout=0.5)
    
    video_ckpt = torch.load('best_video_model.pth', map_location=device, weights_only=False)
    video_model.load_state_dict(video_ckpt['model_state_dict'])
    print(f"‚úÖ Video Model charg√© (epoch {video_ckpt['epoch']+1})")
    
    ef_ckpt = torch.load('best_ef_model.pth', map_location=device, weights_only=False)
    ef_model.load_state_dict(ef_ckpt['model_state_dict'])
    print(f"‚úÖ EF Model charg√© (epoch {ef_ckpt['epoch']+1})\n")
    
    # Fusion model
    fusion_model = OptimizedMultimodalFusion(
        video_model, ef_model, num_classes=2, dropout=0.5
    ).to(device)
    
    total_params = sum(p.numel() for p in fusion_model.parameters())
    print(f"üß† Fusion Model: {total_params:,} param√®tres\n")
    
    # Geler branches
    fusion_model.freeze_branches()
    print("‚úÖ Branches gel√©es (video + EF)\n")
    
    # Class weights
    raw_weights = config['class_weights']
    class_weights = torch.tensor([
        float(raw_weights['0']) * 2.5,
        float(raw_weights['1']) * 0.8
    ], dtype=torch.float32, device=device)
    
    # Optimizer
    optimizer = optim.AdamW([
        {'params': fusion_model.cross_attention.parameters(), 'lr': 1e-3},
        {'params': fusion_model.fusion_classifier.parameters(), 'lr': 1e-3}
    ], weight_decay=1e-4)
    
    scheduler = CosineAnnealingWarmRestarts(optimizer, T_0=10, eta_min=1e-6)
    
    # Entra√Ænement
    PHASE1_EPOCHS = 10
    best_balance_score = 0.0
    best_f1 = 0.0
    history = []
    
    print("üöÄ D√©but de l'entra√Ænement Phase 1...\n")
    start_time = time.time()
    
    for epoch in range(PHASE1_EPOCHS):
        # Lambda progressif
        if epoch < 4:
            lambda_attn = 0.2
            phase_name = "Exploration"
        elif epoch < 7:
            lambda_attn = 0.5
            phase_name = "√âquilibrage doux"
        else:
            lambda_attn = 1.0
            phase_name = "√âquilibrage fort"
        
        criterion = ProgressiveMultimodalLoss(
            alpha=class_weights, gamma=2.0,
            fusion_weight=0.7, branch_weight=0.15,
            lambda_attn_reg=lambda_attn
        )
        
        epoch_start = time.time()
        print(f"Epoch {epoch+1}/{PHASE1_EPOCHS} - {phase_name} (Œª={lambda_attn})")
        print("-" * 70)
        
        # Train
        train_losses, train_acc, train_attn = train_epoch(
            fusion_model, train_loader, criterion, optimizer, device
        )
        
        # Validation
        val_losses, val_acc, val_f1, val_auc, r_unhealthy, r_healthy, val_attn = validate_epoch(
            fusion_model, val_loader, criterion, device
        )
        
        scheduler.step()
        
        # Balance score
        balance_penalty = abs(0.5 - val_attn[0].item())
        balance_score = 1.0 - balance_penalty
        
        # Historique
        history.append({
            'epoch': epoch + 1,
            'lambda_attn': lambda_attn,
            'train_loss': train_losses['total'],
            'train_acc': train_acc,
            'val_loss': val_losses['total'],
            'val_acc': val_acc,
            'val_f1': val_f1,
            'val_auc': val_auc,
            'recall_unhealthy': r_unhealthy,
            'recall_healthy': r_healthy,
            'balance_score': balance_score,
            'video_attn': val_attn[0].item(),
            'ef_attn': val_attn[1].item(),
            'attn_reg_loss': train_losses['attn_reg']
        })
        
        # Affichage
        print(f"Train - Loss: {train_losses['total']:.4f} | Acc: {train_acc:.4f}")
        print(f"  Attention: Video={train_attn[0]:.3f}, EF={train_attn[1]:.3f}")
        print(f"  Attn Reg Loss: {train_losses['attn_reg']:.4f}")
        print(f"Val   - Loss: {val_losses['total']:.4f} | Acc: {val_acc:.4f} | F1: {val_f1:.4f}")
        print(f"  AUC: {val_auc:.4f}")
        print(f"  Recalls: Unhealthy={r_unhealthy:.3f}, Healthy={r_healthy:.3f}")
        print(f"  Attention: Video={val_attn[0]:.3f}, EF={val_attn[1]:.3f}")
        print(f"  Balance Score: {balance_score:.3f}")
        
        # Sauvegarder meilleur mod√®le
        is_improvement = (balance_score > best_balance_score and val_f1 >= 0.80) or \
                        (val_f1 > best_f1 and balance_score > 0.70)
        
        if is_improvement:
            best_balance_score = max(balance_score, best_balance_score)
            best_f1 = max(val_f1, best_f1)
            
            torch.save({
                'epoch': epoch,
                'model_state_dict': fusion_model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'scheduler_state_dict': scheduler.state_dict(),
                'balance_score': balance_score,
                'val_f1': val_f1,
                'val_acc': val_acc,
                'val_auc': val_auc,
                'recall_unhealthy': r_unhealthy,
                'recall_healthy': r_healthy,
                'video_attn': val_attn[0].item(),
                'ef_attn': val_attn[1].item(),
                'lambda_attn': lambda_attn
            }, 'best_fusion_phase1_final.pth')
            print(f"‚≠ê Meilleur mod√®le sauvegard√©!")
        
        epoch_time = (time.time() - epoch_start) / 60
        print(f"‚è±Ô∏è  Epoch time: {epoch_time:.1f}min\n")
    
    total_time = (time.time() - start_time) / 60
    print(f"\n{'='*80}")
    print(f"‚úÖ PHASE 1 TERMIN√âE")
    print(f"   Temps total: {total_time:.1f} min ({total_time/60:.2f}h)")
    print(f"   Best Balance Score: {best_balance_score:.3f}")
    print(f"   Best F1: {best_f1:.4f}")
    print(f"{'='*80}\n")
    
    return fusion_model, history


# ============================================================================
# 7. MAIN - EX√âCUTION
# ============================================================================

if __name__ == "__main__":
    print("\n" + "="*80)
    print("üöÄ D√âMARRAGE PARTIE 1 - PHASE 1")
    print("="*80 + "\n")
    
    # V√©rifier pr√©requis
    print("üì¶ V√©rification des pr√©requis...")
    required_loaders = ['train_loader', 'val_loader']
    missing = [obj for obj in required_loaders if obj not in globals()]
    
    if missing:
        print(f"‚ùå ERREUR: {missing} manquants")
        print("üí° Ex√©cute d'abord le preprocessing")
        raise RuntimeError("Preprocessing requis!")
    
    files = ['best_video_model.pth', 'best_ef_model.pth', 'preprocessing_config_v2.json']
    for f in files:
        if not os.path.exists(f):
            print(f"‚ùå {f} manquant!")
            raise FileNotFoundError(f)
    
    print("‚úÖ Tous les pr√©requis OK\n")
    
    # Charger config
    with open('preprocessing_config_v2.json', 'r') as f:
        config = json.load(f)
    
    # Entra√Ænement Phase 1
    fusion_model, history = train_phase1(train_loader, val_loader, device, config)
    
    # Sauvegarder historique
    print("="*80)
    print("üíæ SAUVEGARDE FINALE PARTIE 1")
    print("="*80 + "\n")
    
    with open('fusion_phase1_history.json', 'w') as f:
        json.dump(history, f, indent=4)
    print("‚úÖ Historique: fusion_phase1_history.json")
    
    # Config pour Phase 2
    phase1_config = {
        'completed': True,
        'best_balance_score': max([h['balance_score'] for h in history]),
        'best_f1': max([h['val_f1'] for h in history]),
        'final_video_attn': history[-1]['video_attn'],
        'final_ef_attn': history[-1]['ef_attn'],
        'total_epochs': len(history)
    }
    
    with open('fusion_phase1_config.json', 'w') as f:
        json.dump(phase1_config, f, indent=4)
    print("‚úÖ Config Phase 2: fusion_phase1_config.json")
    
    # Visualisation rapide
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    epochs = [h['epoch'] for h in history]
    
    # Attention Evolution
    ax1 = axes[0]
    ax1.plot(epochs, [h['video_attn'] for h in history], 
             label='Video', marker='o', linewidth=2, color='#E74C3C')
    ax1.plot(epochs, [h['ef_attn'] for h in history], 
             label='EF', marker='s', linewidth=2, color='#3498DB')
    ax1.axhline(y=0.5, color='green', linestyle='--', label='Target')
    ax1.set_xlabel('Epoch', fontweight='bold')
    ax1.set_ylabel('Attention Weight', fontweight='bold')
    ax1.set_title('Phase 1 - Attention √âvolution', fontweight='bold')
    ax1.legend()
    ax1.grid(alpha=0.3)
    
    # Balance Score
    ax2 = axes[1]
    ax2.plot(epochs, [h['balance_score'] for h in history], 
             marker='o', linewidth=2.5, color='purple')
    ax2.set_xlabel('Epoch', fontweight='bold')
    ax2.set_ylabel('Balance Score', fontweight='bold')
    ax2.set_title('Phase 1 - Balance Score', fontweight='bold')
    ax2.grid(alpha=0.3)
    
    # F1 Score
    ax3 = axes[2]
    ax3.plot(epochs, [h['val_f1'] for h in history], 
             marker='o', linewidth=2.5, color='#9B59B6')
    ax3.set_xlabel('Epoch', fontweight='bold')
    ax3.set_ylabel('F1 Score', fontweight='bold')
    ax3.set_title('Phase 1 - Validation F1', fontweight='bold')
    ax3.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('fusion_phase1_results.png', dpi=300, bbox_inches='tight')
    print("‚úÖ Visualisation: fusion_phase1_results.png")
    plt.show()
    
    # ZIP pour t√©l√©chargement
    print("\nüì¶ Cr√©ation du package Phase 1...")
    with zipfile.ZipFile('/kaggle/working/fusion_phase1_package.zip', 'w', zipfile.ZIP_DEFLATED) as zipf:
        files_to_zip = [
            'best_fusion_phase1_final.pth',
            'fusion_phase1_history.json',
            'fusion_phase1_config.json',
            'fusion_phase1_results.png'
        ]
        for f in files_to_zip:
            if os.path.exists(f):
                zipf.write(f, f)
                print(f"   ‚úì {f}")
    
    size_mb = os.path.getsize('/kaggle/working/fusion_phase1_package.zip') / (1024**2)
    print(f"\n‚úÖ Package cr√©√©: fusion_phase1_package.zip ({size_mb:.2f} MB)")
    
    # R√©sum√© final
    print("\n" + "="*80)
    print("üìä R√âSUM√â PHASE 1")
    print("="*80)
    
    best_epoch = max(history, key=lambda x: x['balance_score'])
    print(f"\nüèÜ Meilleure epoch: {best_epoch['epoch']}")
    print(f" Balance Score: {best_epoch['balance_score']:.3f}")
    print(f" F1 Score: {best_epoch['val_f1']:.4f}")
    print(f" AUC: {best_epoch['val_auc']:.4f}")
    print(f" Accuracy: {best_epoch['val_acc']:.4f}")
    print(f" Recall Unhealthy: {best_epoch['recall_unhealthy']:.3f}")
    print(f" Recall Healthy: {best_epoch['recall_healthy']:.3f}")
    print(f" Attention finale - Video: {best_epoch['video_attn']:.3f} | EF: {best_epoch['ef_attn']:.3f}")

    print("\nüéâ PHASE 1 COMPL√âT√âE AVEC SUCC√àS !")
    print("Tu peux maintenant passer √† la Partie 2/2 - Phase 2 (fine-tuning complet).")
    print("üì¶ T√©l√©charge le package : fusion_phase1_package.zip")
    print("="*80)
```

```python
"""
√âVALUATION TEST SET - PHASE 1 FUSION (VERSION COMPL√àTE)
Inclut les d√©finitions des architectures + √©valuation
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.models.video as video_models
from sklearn.metrics import (accuracy_score, f1_score, precision_score,
                             recall_score, roc_auc_score, confusion_matrix,
                             classification_report, roc_curve)
from tqdm.auto import tqdm
import json
import os
import warnings
warnings.filterwarnings('ignore')

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("="*80)
print("üìä √âVALUATION TEST SET - PHASE 1 FUSION")
print("="*80 + "\n")

# ============================================================================
# D√âFINITIONS DES ARCHITECTURES (COPIE DEPUIS FUSION)
# ============================================================================

class SpatialAttention(nn.Module):
    def __init__(self, in_channels):
        super().__init__()
        self.conv = nn.Conv3d(in_channels + 1, 1, kernel_size=1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, features, masks):
        if masks.shape[-2:] != features.shape[-2:]:
            masks = F.interpolate(
                masks.view(masks.size(0), 1, -1, masks.size(-2), masks.size(-1)),
                size=features.shape[-3:], mode='trilinear', align_corners=False
            )
        combined = torch.cat([features, masks], dim=1)
        attention = self.sigmoid(self.conv(combined))
        return features * attention


class VideoModelWithSegmentation(nn.Module):
    def __init__(self, num_classes=2, pretrained=True, dropout=0.5):
        super().__init__()
        self.backbone = video_models.r2plus1d_18(pretrained=pretrained)
        self.backbone.stem[0] = nn.Conv3d(1, 45, kernel_size=(1,7,7),
                                          stride=(1,2,2), padding=(0,3,3), bias=False)
        self.attention = SpatialAttention(in_channels=64)
        in_features = self.backbone.fc.in_features
        self.backbone.fc = nn.Identity()
        
        self.classifier = nn.Sequential(
            nn.Linear(in_features, 256), nn.ReLU(), nn.BatchNorm1d(256), nn.Dropout(dropout),
            nn.Linear(256, 128), nn.ReLU(), nn.BatchNorm1d(128), nn.Dropout(dropout*0.5),
            nn.Linear(128, num_classes)
        )
        
        self.intermediate_features = None
        def hook_fn(module, input, output):
            self.intermediate_features = output
        self.backbone.layer1.register_forward_hook(hook_fn)
    
    def forward(self, video, masks, return_features=False):
        x = video.permute(0, 2, 1, 3, 4)
        masks = masks.permute(0, 2, 1, 3, 4)
        x = self.backbone.stem(x)
        x = self.backbone.layer1(x)
        x = self.attention(x, masks)
        x = self.backbone.layer2(x)
        x = self.backbone.layer3(x)
        x = self.backbone.layer4(x)
        x = self.backbone.avgpool(x)
        features = x.flatten(1)
        logits = self.classifier(features)
        if return_features:
            return logits, features
        return logits


class EFBranch(nn.Module):
    def __init__(self, num_classes=2, dropout=0.5):
        super().__init__()
        self.ef_branch = nn.Sequential(
            nn.Linear(1, 64), nn.ReLU(), nn.BatchNorm1d(64), nn.Dropout(dropout),
            nn.Linear(64, 128), nn.ReLU(), nn.BatchNorm1d(128), nn.Dropout(dropout),
            nn.Linear(128, 64), nn.ReLU(), nn.BatchNorm1d(64), nn.Dropout(dropout*0.5),
            nn.Linear(64, num_classes)
        )
    
    def forward(self, ef, return_features=False):
        ef = ef.unsqueeze(1)
        if return_features:
            x = ef
            for i, layer in enumerate(self.ef_branch):
                x = layer(x)
                if i == len(self.ef_branch) - 2:
                    features = x
            logits = self.ef_branch[-1](features)
            return logits, features
        return self.ef_branch(ef)


class SoftBalancedCrossModalAttention(nn.Module):
    def __init__(self, video_dim=512, ef_dim=64, hidden_dim=128):
        super().__init__()
        self.video_proj = nn.Linear(video_dim, hidden_dim)
        self.ef_proj = nn.Linear(ef_dim, hidden_dim)
        self.attention_score = nn.Linear(hidden_dim, 1)
    
    def forward(self, video_features, ef_features):
        video_proj = self.video_proj(video_features)
        ef_proj = self.ef_proj(ef_features)
        
        video_score = self.attention_score(torch.tanh(video_proj))
        ef_score = self.attention_score(torch.tanh(ef_proj))
        
        scores = torch.cat([video_score, ef_score], dim=1)
        attention_weights = F.softmax(scores, dim=1)
        
        attended_video = video_features * attention_weights[:, 0:1]
        attended_ef = ef_features * attention_weights[:, 1:2]
        
        return attended_video, attended_ef, attention_weights


class OptimizedMultimodalFusion(nn.Module):
    def __init__(self, video_model, ef_model, num_classes=2, dropout=0.5):
        super().__init__()
        self.video_branch = video_model
        self.ef_branch = ef_model
        self.cross_attention = SoftBalancedCrossModalAttention(512, 64, 128)
        
        self.fusion_classifier = nn.Sequential(
            nn.Linear(576, 256), nn.ReLU(), nn.BatchNorm1d(256), nn.Dropout(dropout),
            nn.Linear(256, 128), nn.ReLU(), nn.BatchNorm1d(128), nn.Dropout(dropout*0.5),
            nn.Linear(128, 64), nn.ReLU(), nn.BatchNorm1d(64), nn.Dropout(dropout*0.25),
            nn.Linear(64, num_classes)
        )
    
    def forward(self, video, masks, ef):
        video_logits, video_features = self.video_branch(video, masks, return_features=True)
        ef_logits, ef_features = self.ef_branch(ef, return_features=True)
        
        attended_video, attended_ef, attention_weights = self.cross_attention(
            video_features, ef_features
        )
        
        fused_features = torch.cat([attended_video, attended_ef], dim=1)
        fusion_logits = self.fusion_classifier(fused_features)
        
        return fusion_logits, video_logits, ef_logits, attention_weights


print("‚úÖ Architectures d√©finies\n")

# ============================================================================
# CHARGEMENT DU MOD√àLE
# ============================================================================
print("üì• Chargement du mod√®le Phase 1...")

# Chemins
FUSION_MODEL_PATH = '/kaggle/input/modele-fusion/other/default/1/best_fusion_phase1_final.pth'
VIDEO_MODEL_PATH = 'best_video_model.pth'  # Si disponible localement
EF_MODEL_PATH = 'best_ef_model.pth'  # Si disponible localement

# Alternative: si les mod√®les sont aussi dans input
if not os.path.exists(VIDEO_MODEL_PATH):
    VIDEO_MODEL_PATH = '/kaggle/input/video-model-package/best_video_model.pth'
if not os.path.exists(EF_MODEL_PATH):
    EF_MODEL_PATH = '/kaggle/input/ef-model-package/best_ef_model.pth'

# V√©rifications
if not os.path.exists(FUSION_MODEL_PATH):
    print(f"‚ùå ERREUR: Mod√®le fusion non trouv√©: {FUSION_MODEL_PATH}")
    raise FileNotFoundError(f"Mod√®le fusion manquant")

# Charger checkpoint fusion
checkpoint = torch.load(FUSION_MODEL_PATH, map_location=device, weights_only=False)
print(f"‚úÖ Checkpoint fusion charg√© (epoch {checkpoint['epoch']+1})")

# Cr√©er les branches
print("üîß Cr√©ation des branches vid√©o et EF...")
video_model = VideoModelWithSegmentation(num_classes=2, pretrained=False, dropout=0.5)
ef_model = EFBranch(num_classes=2, dropout=0.5)

# Charger les poids des branches si disponibles (optionnel pour l'√©valuation)
# Les poids sont d√©j√† dans le checkpoint fusion
print("‚úÖ Branches cr√©√©es")

# Cr√©er le mod√®le fusion
print("üîß Cr√©ation du mod√®le fusion...")
fusion_model = OptimizedMultimodalFusion(
    video_model, ef_model, num_classes=2, dropout=0.5
).to(device)

# Charger les poids
fusion_model.load_state_dict(checkpoint['model_state_dict'])
fusion_model.eval()

print(f"‚úÖ Mod√®le fusion charg√© et pr√™t")
print(f"   Val F1: {checkpoint['val_f1']:.4f}")
print(f"   Balance Score: {checkpoint['balance_score']:.3f}")
print(f"   Video Attn: {checkpoint['video_attn']:.3f}")
print(f"   EF Attn: {checkpoint['ef_attn']:.3f}")
print()

# ============================================================================
# √âVALUATION TEST SET
# ============================================================================
print("="*80)
print("üß™ √âVALUATION SUR TEST SET")
print("="*80)

# V√©rifier test_loader
if 'test_loader' not in globals():
    print("\n‚ùå ERREUR: test_loader non trouv√©")
    print("üí° SOLUTION:")
    print("   1. Ex√©cute d'abord le code de preprocessing (Document 1)")
    print("   2. Puis relance ce code d'√©valuation")
    raise RuntimeError("test_loader manquant")

print(f"‚úì Test loader: {len(test_loader)} batches\n")

# √âvaluation
all_preds = []
all_labels = []
all_probs = []
all_video_probs = []
all_ef_probs = []
all_attention = []

print("üîÑ √âvaluation en cours...")
with torch.no_grad():
    for batch in tqdm(test_loader, desc='Testing'):
        videos = batch['video'].to(device)
        masks = batch['masks'].to(device)
        ef = batch['ef'].to(device)
        labels = batch['label'].to(device)
        
        # Forward pass
        fusion_logits, video_logits, ef_logits, attn = fusion_model(videos, masks, ef)
        
        # Pr√©dictions avec seuil 0.15
        probs = torch.softmax(fusion_logits, dim=1)
        preds = (probs[:, 1] > 0.15).long()
        
        # Probabilit√©s branches individuelles
        video_probs = torch.softmax(video_logits, dim=1)[:, 1]
        ef_probs = torch.softmax(ef_logits, dim=1)[:, 1]
        
        # Sauvegarder
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_probs.extend(probs[:, 1].cpu().numpy())
        all_video_probs.extend(video_probs.cpu().numpy())
        all_ef_probs.extend(ef_probs.cpu().numpy())
        all_attention.append(attn.cpu())
        
        del videos, masks, ef, labels
        torch.cuda.empty_cache()

# Convertir
all_preds = np.array(all_preds)
all_labels = np.array(all_labels)
all_probs = np.array(all_probs)
all_video_probs = np.array(all_video_probs)
all_ef_probs = np.array(all_ef_probs)
avg_attention = torch.cat(all_attention, dim=0).mean(dim=0)

print("‚úÖ √âvaluation termin√©e\n")

# ============================================================================
# CALCUL M√âTRIQUES
# ============================================================================
print("="*80)
print("üìà R√âSULTATS TEST SET - FUSION PHASE 1")
print("="*80 + "\n")

# M√©triques globales
test_acc = accuracy_score(all_labels, all_preds)
test_f1 = f1_score(all_labels, all_preds, average='weighted')
test_precision = precision_score(all_labels, all_preds, average='weighted')
test_recall = recall_score(all_labels, all_preds, average='weighted')
test_auc = roc_auc_score(all_labels, all_probs)

# Confusion matrix
cm = confusion_matrix(all_labels, all_preds)
recall_unhealthy = cm[0, 0] / (cm[0, 0] + cm[0, 1]) if (cm[0, 0] + cm[0, 1]) > 0 else 0
recall_healthy = cm[1, 1] / (cm[1, 0] + cm[1, 1]) if (cm[1, 0] + cm[1, 1]) > 0 else 0

# Affichage
print(f"üéØ M√âTRIQUES PRINCIPALES:")
print(f"   Accuracy:  {test_acc:.4f} ({test_acc*100:.2f}%)")
print(f"   F1 Score:  {test_f1:.4f}")
print(f"   Precision: {test_precision:.4f}")
print(f"   Recall:    {test_recall:.4f}")
print(f"   AUC-ROC:   {test_auc:.4f}")
print()

print(f"üìä RECALLS PAR CLASSE:")
print(f"   Unhealthy: {recall_unhealthy:.3f} ({recall_unhealthy*100:.1f}%)")
print(f"   Healthy:   {recall_healthy:.3f} ({recall_healthy*100:.1f}%)")
print()

print(f"‚öñÔ∏è  CONTRIBUTION DES MODALIT√âS:")
print(f"   Video: {avg_attention[0]:.3f} ({avg_attention[0]*100:.1f}%)")
print(f"   EF:    {avg_attention[1]:.3f} ({avg_attention[1]*100:.1f}%)")
print()

print("üìã CLASSIFICATION REPORT:")
print(classification_report(all_labels, all_preds, 
                          target_names=['Unhealthy', 'Healthy'], 
                          digits=4))

# ============================================================================
# COMPARAISON BRANCHES
# ============================================================================
print("\n" + "="*80)
print("üîç COMPARAISON: FUSION vs BRANCHES INDIVIDUELLES")
print("="*80 + "\n")

# √âvaluer branches
video_preds = (all_video_probs > 0.15).astype(int)
ef_preds = (all_ef_probs > 0.15).astype(int)

video_acc = accuracy_score(all_labels, video_preds)
ef_acc = accuracy_score(all_labels, ef_preds)
video_f1 = f1_score(all_labels, video_preds, average='weighted')
ef_f1 = f1_score(all_labels, ef_preds, average='weighted')

print(f"{'Mod√®le':<20} {'Accuracy':<12} {'F1 Score':<12} {'Am√©lioration':<15}")
print("-" * 60)
print(f"{'Video seul':<20} {video_acc:.4f}      {video_f1:.4f}")
print(f"{'EF seul':<20} {ef_acc:.4f}      {ef_f1:.4f}")
print(f"{'FUSION Phase 1':<20} {test_acc:.4f}      {test_f1:.4f}      "
      f"{'üåü +' + f'{(test_acc - max(video_acc, ef_acc))*100:.2f}%'}")
print()

# ============================================================================
# DIAGNOSTIC
# ============================================================================
print("="*80)
print("üîç DIAGNOSTIC FINAL")
print("="*80 + "\n")

if test_auc >= 0.95 and recall_unhealthy >= 0.85 and recall_healthy >= 0.85:
    print("üåü EXCELLENT ! Ton mod√®le est pr√™t pour la production.")
    print("   ‚úÖ AUC √©lev√© (>0.95)")
    print("   ‚úÖ Recalls √©quilibr√©s (>85%)")
    print("   ‚úÖ Fusion efficace\n")
    print("üí° RECOMMANDATION:")
    print("   ‚Üí Tu peux t'arr√™ter √† la Phase 1 !")
    print("   ‚Üí La Phase 2 n'est PAS n√©cessaire.")
elif test_acc >= 0.90 and abs(recall_unhealthy - recall_healthy) < 0.10:
    print("‚úÖ TR√àS BON ! Mod√®le performant et √©quilibr√©.")
    print("   ‚úÖ Accuracy >90%")
    print("   ‚úÖ Recalls √©quilibr√©s\n")
    print("üí° RECOMMANDATION:")
    print("   ‚Üí Phase 1 suffit largement")
    print("   ‚Üí Phase 2 optionnelle (gain marginal)")
elif test_acc >= 0.85:
    print("‚úì BON. Mod√®le correct mais peut √™tre am√©lior√©.")
    print(f"   ‚ö†Ô∏è  Recall Unhealthy: {recall_unhealthy:.1%}")
    print(f"   ‚ö†Ô∏è  Recall Healthy: {recall_healthy:.1%}\n")
    print("üí° RECOMMANDATION:")
    print("   ‚Üí Consid√®re la Phase 2 pour am√©liorer")
else:
    print("‚ö†Ô∏è  PERFORMANCES MOD√âR√âES")
    print(f"   Accuracy: {test_acc:.1%}\n")
    print("üí° RECOMMANDATION:")
    print("   ‚Üí Phase 2 RECOMMAND√âE")

# Overfitting check
val_f1 = checkpoint['val_f1']
gap = val_f1 - test_f1

print(f"\nüìâ G√âN√âRALISATION:")
print(f"   Val F1:  {val_f1:.4f}")
print(f"   Test F1: {test_f1:.4f}")
print(f"   Gap:     {gap:.4f}")

if gap > 0.05:
    print("   ‚ö†Ô∏è  Overfitting d√©tect√© (gap >5%)")
elif gap > 0.02:
    print("   ‚ö†Ô∏è  L√©ger overfitting (gap 2-5%)")
else:
    print("   ‚úÖ Bonne g√©n√©ralisation (gap <2%)")

# Sauvegarde
results = {
    'test_accuracy': float(test_acc),
    'test_f1': float(test_f1),
    'test_auc': float(test_auc),
    'recall_unhealthy': float(recall_unhealthy),
    'recall_healthy': float(recall_healthy),
    'attention_video': float(avg_attention[0]),
    'attention_ef': float(avg_attention[1]),
    'comparison': {
        'video_acc': float(video_acc),
        'ef_acc': float(ef_acc),
        'fusion_acc': float(test_acc),
        'improvement': float(test_acc - max(video_acc, ef_acc))
    }
}

with open('fusion_phase1_test_results.json', 'w') as f:
    json.dump(results, f, indent=4)

print("\nüíæ R√©sultats sauvegard√©s: fusion_phase1_test_results.json")
print("="*80)
```
